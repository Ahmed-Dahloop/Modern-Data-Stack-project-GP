```python
expected_tables = [
    "customers",
    "geolocation",
    "orders",
    "order_items",
    "order_payments",
    "order_reviews",
    "products",
    "sellers",
    "category_translation"
]

existing_tables = [
    row.tableName
    for row in spark.sql("SHOW TABLES IN olist.ods").collect()
]

missing_tables = []

for table in expected_tables:
    if table not in existing_tables:
        missing_tables.append(table)

if len(missing_tables) == 0:
    print("All ODS tables exist.")
else:
    raise Exception(f"Missing ODS tables: {missing_tables}")
```
